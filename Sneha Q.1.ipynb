For this question, you are required to analyse the iris dataset (iris.csv) using Python.
```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(data["PetalLengthCm"])
```

```python
plt.hist(data["PetalWidthCm"])
```

```python
plt.hist(data["SepalLengthCm"])
```

```python
plt.hist(data["SepalWidthCm"])
```

```python
sns.histplot(data["PetalWidthCm"],kde=True)
sns.histplot(data["PetalLengthCm"],kde=True)
sns.histplot(data["SepalWidthCm"],kde=True)
sns.histplot(data["SepalLengthCm"],kde=True)
```

```python
plt.boxplot(data["SepalLengthCm"])
```

```python
plt.boxplot(data["SepalWidthCm"])
```

```python
plt.scatter(data["SepalLengthCm"],data["SepalWidthCm"])
```

```python
print(data["SepalLengthCm"].mean())
print(data["SepalWidthCm"].mean())
print(data["PetalLengthCm"].mean())
print(data["PetalWidthCm"].mean())
```

```python
print(data["SepalLengthCm"].median())
print(data["SepalWidthCm"].median())
print(data["PetalLengthCm"].median())
print(data["PetalWidthCm"].median())
```

```python
print(data["SepalLengthCm"].mode())
print(data["SepalWidthCm"].mode())
print(data["PetalLengthCm"].mode())
print(data["PetalWidthCm"].mode())
```
